# Problem 1: Set Operations

# Define two sets, A and B, with at least 5 elements each.
# Write a function to find the union of sets A and B.
# Write a function to find the intersection of sets A and B.
#Write a function to find the difference between sets A and B (elements in A but not in B).
# Write a function to check if set A is a subset of set B.
# Write a function to find the symmetric difference between sets A and B.
# Write a function to remove duplicates from a list and convert it into a set.
# Write a function to determine if two sets have any elements in common.
# Write a function to check if a set is empty.
# Write a function to remove a specific element from a set.
# Write a function to check if two sets are disjoint (have no elements in common).
# Write a function to find the Cartesian product of two sets.
# Write a function to generate the power set of a set.
# Write a function to find the maximum and minimum elements of a set.
# Write a function to calculate the Jaccard similarity coefficient between two sets.


```python
def union(a,b):
  return a | b

def instruction(a,b):
  return a & b

def difference(a,b):
  return a - b

def subset(a,b):
  return a <= b

def symmetric_difference(a,b):
  return a ^ b

def remove_duplicates(lst):
    x=[]
    for i in lst:
        if i not in x:
            x.append(i)
    return set(x)

def common(a,b):

  if  a.isdisjoint(b):
    return False
  else:
    return True

def is_empty(a):
  if len(a)==0:
    print("Set Is Empty")
  else:
    print("Set Is NOT Empty")

def remove_element(a,b):
  if b not in a:
    print("Element Not Found")
  else:
    a.remove(b)
    print("Set After Remove :",a)

def is_joint(a,b):
  if a.isdisjoint(b):
    print("Sets Are Joint")
  else:
    print("Sets Are Not Joint")

def power_set(a):
  power_set = [set()]
  for element in a:
    power_set += [subset | {element} for subset in power_set]
  return power_set

def max_min(a):
  print("Maximum Element :",max(a))
  print("Minimum Element :",min(a))

def jaccard_similarity(a,b):
  intersection = len(a & b)
  union = len(a | b)
  return intersection / union

def cartesian_product(a,b):
  return [(x,y) for x in a for y in b]

a = {1,2,3,4,5,6}
b={2,3,4,20,22,30,25}
l=[1,1,2,3,4,5,6,6,7,8]

print("Union =",union(a,b))
print("Intersection =",instruction(a,b))
print("Difference =",difference(a,b))
print("Subset =",subset(a,b))
print("Symmetric Difference =",symmetric_difference(a,b))
print("Remove Duplicates =",remove_duplicates(l))
print("Common is",common(a,b))
is_empty(a)
remove_element(a,3)
is_joint(a,b)
max_min(a)
print("Power Set =",power_set(a))
print("Jaccard Similarity =",jaccard_similarity(a,b))
print("Cartesian Product =",cartesian_product(a,b))
```

# Problem 3: Tuple Operations

# Create a tuple containing the names of 10 programming languages.
# Write a function to reverse the elements of a tuple.
# Write a function to find the index of a specific element in a tuple.
# Write a function to convert a tuple to a list.
# Write a function to count the occurrences of a specific element in a tuple.

# Write a function to find the average value of a tuple of numbers.
# Write a function to flatten a tuple of tuples into a single tuple.
# Write a function to find the index of the first occurrence of a tuple within another tuple.

```python
def reverse(t):
  return t[::-1]

def index(t,x):
  return t.index(x)

def convert(t):
  return list(t)

def count(t,x):
  return t.count(x)

def length(t):
  return len(t)

def equal(t1,t2):
  if t1 == t2:
    print("The Two Is Equal")
  else:
    print("Not Equal")

def to_string(t):
  return str(t)

def sort(t):
  return sorted(t)

def zib(t1,t2):
  return zip(t1,t2)

def sliccing(t,x,y):
  return t [x:y]

def average(t):
  return sum(t)/len(t)

def flatten(t):
  result = []
  for item in t:
    if isinstance(item, tuple):
      result.extend(flatten(item))
    else:
      result.append(item)
  return tuple(result)

def first_occurrence(t1,t2):
  for i in range(len(t1)):
    if t1[i:i+len(t2)] == t2:
      return i
  return -1

t =('python','html','css','c','c++','java','javascript','php','c#','SQL')
t2 =(2,1,4,3,0,4,5,6,7,8,9,10)

print("Reversed Tuble :",reverse(t))
print("The Index Of 'Java' Is :",index(t,"java"))
convert(t)
print("Count Of Java :",count(t,'java'))
print("The Length Of Tuble =",length(t))
equal(t,t2)
to_string(t2)
print("Sorted Tuble :",sort(t2))
print("Zipped Tuples :", list(zib(t,t2)))
to_string(t)
print('Slice:',sliccing(t2,0,4))
print("Average Of Tuble :",average(t2))
print("Flatten Tuble :",flatten(t2))
print("First Occurrence :",first_occurrence(t2,t))
```
